```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/loan_data.csv')
df.head()
```

```python
# Get a concise summary of the dataframe
df.info()
# Get descriptive statistics for numerical columns
df.describe()
```

```python
# Create a count plot to see the distribution of the target variable
plt.figure(figsize=(8, 5))
sns.countplot(x='not.fully.paid', data=df, palette=['#43a047', '#e53935'])
plt.title('Distribution of Loan Repayment Status', fontsize=16)
plt.xlabel('Loan Not Fully Paid (1 = Yes, 0 = No)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Print the exact numbers and percentage
print("Repayment Status Counts:")
print(df['not.fully.paid'].value_counts())
print("\nPercentage of Defaults:")
print(f"{df['not.fully.paid'].value_counts(normalize=True)[1]*100:.2f}%")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='purpose', data=df, hue='not.fully.paid', palette=['#43a047', '#e53935'])
plt.title('Loan Purpose vs. Repayment Status', fontsize=16)
plt.xlabel('Purpose of Loan', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right') # ha='right' aligns the labels better
plt.tight_layout() # Adjusts plot to prevent labels from overlapping
plt.show()
```

```python
+plt.figure(figsize=(10, 6))
df[df['not.fully.paid']==0]['fico'].hist(alpha=0.6, color='blue', bins=35, label='Paid Fully')
df[df['not.fully.paid']==1]['fico'].hist(alpha=0.6, color='red', bins=35, label='Not Paid Fully')
plt.legend()
plt.xlabel('FICO Score', fontsize=12)
plt.title('FICO Score Distribution by Repayment Status', fontsize=16)
plt.show()
```

```python
# Using seaborn's jointplot to see the relationship and distributions
sns.jointplot(x='fico', y='int.rate', data=df, color='purple', kind='hex')
plt.show()
```

```python
sns.boxplot(x='not.fully.paid', y='log.annual.inc', data=df)
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='viridis',
    linewidths=1,
    linecolor='black'
)
plt.show()
```
